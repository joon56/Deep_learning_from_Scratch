```python
import numpy as np
x = np.array([-1.0, 1.0, 2.0])
y = x > 0
y # 이렇게 y를 출력하면 bool형의 배열이 만들어진다.
```

```python
y2 = y.astype(np.int) # y가 bool형이니까 이걸 다시 int로 바꾸면 false는 0, true는 1임
y2
```

```python
# 계단 함수
import numpy as np
import matplotlib.pylab as plt

def step_func(x):
    return np.array(x > 0, dtype=np.int)

x = np.arange(-5.0, 5.0, 0.1)
y = step_func(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1) # y축의 범위 지정
plt.show()
```

```python
# 시그모이드 함수
def sigmoid (x):
    return 1 / (1 + np.exp(-x))

x = np.array([-1.0, 1.0, 2.0])
sigmoid(x)
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.show()
```

```python
# ReLU 함수
def relu(x):
    return np.maximum(0, x) # 넘파이 maximum은 (앞, 뒤) 앞과 뒤 중 더 큰 값을 선택해서 반환하는 함수

x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)
plt.plot(x, y)
plt.ylim(-1.0, 5.0)
plt.show()
```
